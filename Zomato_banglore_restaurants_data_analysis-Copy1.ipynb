```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df=pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis=1, inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)#removing duplicate rows from datasets
```

```python
df.shape
```

```python
#cleaning rate values
df['rate'].unique()
```

```python
def handlerate(value):
    if value=="NEW" or value=='-':
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'costfor2','listed_in(type)':'type'},inplace=True)
```

```python
df.columns
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
df.head()
```

```python
df['costfor2'].unique()
```

```python
def handlecostfor2(value):
    value=str(value)
    if "," in value:
        value=value.replace(",","")
        return float(value)
    else:
        return float(value)
df['costfor2']=df['costfor2'].apply(handlecostfor2)
df['costfor2'].unique()
```

```python
#working on rest type pf column
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
```

```python
rest_type_lessthan100=rest_type[rest_type<1000]
rest_type_lessthan100
```

```python
#Making Rest Types less than 1000 in frequency as others
def handle_rest_type(value):
    if value in rest_type_lessthan100:
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
```

```python
locationlessthan300=location[location<300]
```

```python
def handle_location(value):
    if value in locationlessthan300:
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
```

```python
cuisineslessthan100=cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if value in cuisineslessthan100:
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
